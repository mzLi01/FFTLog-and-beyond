```python
import numpy as np
from fftlogx import hankel, fftlog

import matplotlib.pyplot as plt
```

```python
k, pk = np.loadtxt('Pk_test', usecols=(0, 1), unpack=True)
N = k.size
print('number of input data points: '+str(N))
ell = 100.
nu = 1.01
myfftlog = fftlog(k, pk, nu=nu, N_extrap_low=1500, N_extrap_high=1500, c_window_width=0.25, N_pad=5000)
```

## Test fftlog

```python
r, Fr = myfftlog.fftlog(ell)
fig = plt.figure(figsize=(8, 4))
fig.suptitle(r'$F(y) = \int_0^{\infty} f(x)j_{\ell}(xy) dx/x, \ell=$%.1f' % (ell))

subfig1 = fig.add_subplot(1, 2, 1)
subfig1.set_xscale('log')
subfig1.set_yscale('log')
subfig1.set_xlabel('x')
subfig1.set_ylabel('f(x)')
subfig1.plot(k, pk)
plt.tight_layout()

subfig2 = fig.add_subplot(1, 2, 2)
subfig2.set_title(r'$\nu=$%.2f' % (nu))
subfig2.set_xscale('log')
subfig2.set_yscale('log')
subfig2.set_xlabel('y')
subfig2.set_ylabel('F(y)')
subfig2.plot(r, Fr, label='fftlog')

plt.legend()
plt.tight_layout()
plt.show()
```

## Test on 2 spherical Bessel functions

$$ F(y) = \int \frac{\mathrm{d}x}{x} f(x) j_{l_1}(xy) j_{l_2}(\beta xy) $$
Here we set $\beta=0.5$, $l_1,l_2=5\pm 1$. We perform the numerical integration in Mathematica as comparation.

```python
l1 = np.array([4, 6, 4, 6])
l2 = np.array([6, 4, 4, 6])
r, Fr = myfftlog.jl1_jl2_neq(l1, l2, 0.5)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for i, ax in enumerate(axes.flatten()):
    r_mma, fr_mma = np.loadtxt(f'data/j{l1[i]}j{l2[i]}_mma.txt', usecols=(0, 1), unpack=True)
    ax.loglog(r[i], Fr[i], label='fftlog')
    ax.plot(r_mma, fr_mma, label='mathematica', ls='none', marker='o', markersize=3)
    ax.set_title(f'$j_{{{l1[i]}}}j_{{{l2[i]}}}$')
    ax.legend()
plt.tight_layout()
```

## Test on two derivative of spherical Bessel function
$$ F(y) = \int \frac{\mathrm{d}x}{x} f(x) j_{l}'(xy) j_{l}'(\beta xy) $$
Utilizing
$$ j_l'(x) = \frac{1}{2l+1} (lj_{l-1}(x) - (l+1)j_{l+1}(x)) $$
Here we set $\beta=0.5,~l=5$. We perform the numerical integration in Mathematica as comparation.

```python
def plot_log(ax, x, y, label='', neg_color='grey', color=None, **kwargs):
    mask = y > 0

    pos_kwargs = kwargs.copy()
    if color is not None:
        pos_kwargs['color'] = color
    ax.plot(x[mask], y[mask], label=label, **pos_kwargs)
    ax.plot(x[~mask], -y[~mask], color=neg_color, **kwargs)
```

```python
from fftlogx.jderiv_interp import djl_djl_neq


l = 5
djdj_interp = djl_djl_neq(myfftlog, 0.5, l, l)

result_mma_r, result_mma_fr = np.loadtxt('data/jd5jd5_mma.txt', usecols=(0, 1), unpack=True)

r = np.logspace(-1, 4, 1000)
fr = djdj_interp(r, [l])[0]
plot_log(plt.gca(), r, fr, neg_color='violet', label='fftlog')
plot_log(plt.gca(), result_mma_r, result_mma_fr, label='mathematica',
         neg_color='red', ls='none', marker='o', markersize=6)

plt.xscale('log')
plt.yscale('log')
plt.legend()
```

In the above figure, purple line and red dots represent negative values.
